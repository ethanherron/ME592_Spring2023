#### Classes

```python
class Model(object):
    # class initialization and attributes
    def __init__(self, params, model_type, name='Neural Network'):
        self.name = name
        self.params = params
        self.model_type = model_type 
      
    # internal functions denoted by __.__
    # str 
    def __str__(self):
        return 'Readable string : This is a '+ self.model_type + ' ' + self.name + ' with ' + str(self.params) + ' params'

    # representation
    def __repr__(self):
        return self.model_type + ' ' + self.name + ' with ' + str(self.params) + ' params'
    
    # method
    def printParams(self, multiplier):
        print(self.params*multiplier)
```

```python
m1 = Model(5, 'Supervised', 'LinearRegression')
m1
```

```python
print(m1)
```

```python
m1.name
```

```python
m1.printParams(10)
```

```python
# try changing kernel size and observe what happens
kernel = np.ones((10,10),np.float32)/100 

# arguments are (source_image, depth, kernel)
dst = cv2.filter2D(src=img, ddepth=-1,kernel=kernel)

fig = plt.figure(figsize=(10, 10), dpi=100)
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)

ax1.imshow(img),plt.title('Original')
ax2.imshow(dst),plt.title('Averaging')
plt.show()
```

```python
pic = img[:, :img.shape[0],:]
print(pic.shape)
```

```python
pic = rearrange(pic, 'h w c -> 1 h w c')
print(pic.shape)
```

```python
def img_to_patch(x, patch_size):
    b, c, H, W = x.shape
    x = rearrange(x, 'b c (h nph ph) (w npw pw) -> b (h nph) (w npw) ph pw c', nph=H//patch_size, npw=W//patch_size, ph=patch_size, pw=patch_size)
    x = rearrange(x, 'b (h nph) (w npw) ph pw c -> b (h nph w npw) ph pw c', nph=H//patch_size, npw=W//patch_size)
    return x
```

```python
# einops
patches = img_to_patch(rearrange(pic, 'b w h c -> b c h w'), 60)
print(patches.shape)
```

```python
def plot_patches(patches):
    n_patches = len(patches)
    n_cols = int(math.sqrt(n_patches))
    n_rows = math.ceil(n_patches/n_cols)
    fig, axs = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(20,20))
    axs = axs.reshape(-1)
    for i, patch in enumerate(patches):
        axs[i].imshow(patch)
    plt.tight_layout()
    plt.show()
```

```python
plot_patches(patches.squeeze())
```

```python
from scipy import signal
# can be used to compue moving average
def apply_convolution(sig, window):
    conv = np.repeat([0., 1., 0.], window)
    print('Convolution Kernel:', conv)
    filtered = signal.convolve(sig, conv, mode='same') / window
    return filtered

# apply a median filter, arguments are (volume, kernel)
filtered = signal.medfilt(timedata_np[:,1], 9)
# apply a convolution filter
convolved = apply_convolution(timedata_np[:,1], 9)
plt.plot(filtered)
plt.show()
plt.plot(convolved)
plt.show()
```

## Dimensionality Reduction

```python
from torchvision import datasets
mnist = datasets.MNIST('./data', train=True, download=False)
```

```python
mnist = mnist._load_data()
```

```python
(images, labels) = mnist
print(images.shape, labels.shape)
```

```python
idx = np.random.randint(0, len(images))
plt.imshow(images[idx])
print(labels[idx])
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(images[:10000].reshape(10000,28*28))
```

```python
plt.figure()
plt.scatter(pca_result[:,0], pca_result[:,1], c=labels[:10000])
plt.colorbar()
plt.show()
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
import time
start = time.time()
tsne_results = tsne.fit_transform(images[:10000].reshape(10000,28*28))
end = time.time()
print('Time', end-start)
```

```python
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels[:10000])
plt.colorbar()
plt.show()
```

```python
# applying PCA n high-dimensional data, then using t-SNE can speed up process
pca = PCA(n_components=50)
pca_result_int = pca.fit_transform(images[:10000].reshape(10000,28*28))
start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(pca_result_int)
end = time.time()
print('Time', end-start)
```

```python
plt.figure()
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=labels[:10000])
plt.colorbar()
plt.show()
```
